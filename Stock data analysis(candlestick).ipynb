```python
import yfinance as yf
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import dotenv
from dotenv import load_dotenv
import os

from mpl_finance import candlestick_ohlc
import matplotlib.dates as mpl_dates
```

```python
font_title = {"family":"serif","size":20, "color":"blue"}
font_label = {"family":"serif", "size":15, "color":"darkred"}
```

```python
load_dotenv()
```

```python
ticker_name = os.environ['TICKER_NAME']
```

```python
startDate = datetime.datetime(2024, 1, 1)
```

```python
endDate = datetime.datetime(2024, 4, 30)
```

```python
data = yf.Ticker(ticker_name)
```

```python
meta_df = pd.DataFrame(data.history(start = startDate, end = endDate))
```

```python
meta_df
```

```python

meta_df = meta_df.drop(columns = ['Dividends', 'Stock Splits'])
```

```python
meta_df = meta_df.tail(50).reset_index()
meta_df
```

```python
# print(type(meta_df['Date'].dtypes))
```

```python
plt.figure(figsize=(10,6))

up = meta_df[meta_df.Close >= meta_df.Open]
down = meta_df[meta_df.Close < meta_df.Open]

col1 = 'green'
col2 = 'red'

width = 0.8
width2 = 0.1

plt.bar(up.Date, up.Close-up.Open, width, bottom = up.Open, color = col1)
plt.bar(up.Date, up.High-up.Close, width2, bottom = up.Close, color = col1)
plt.bar(up.Date, up.Low - up.Open, width2 , bottom=up.Open, color = col1)

plt.bar(down.Date, down.Close - down.Open, width, bottom=down.Open, color = col2)
plt.bar(down.Date, down.High - down.Open, width2, bottom= down.Open,  color = col2)
plt.bar(down.Date, down.Low - down.Close, width2, bottom =down.Close, color = col2)

plt.xlabel("Date ", fontdict = font_label)
plt.ylabel("Price", fontdict = font_label)
plt.title("Stock analysis", fontdict= font_title, )

plt.xticks(rotation = 45 , ha = 'right')
plt.grid()
plt.show()
```

# Step Graph

```python
plt.figure(figsize= (10,10))
plt.step(meta_df['Date'], meta_df['Close'], color ="red")


plt.xlabel('Date', fontdict= font_label)
plt.ylabel('Closing Value per day', fontdict= font_label)
plt.title("Closing Values", fontdict=font_title)
plt.xticks(rotation = 45 , ha = 'right')
plt.grid(color = 'skyblue')
plt.show()
```

# Line Graph

```python
plt.figure(figsize=(10,10))
plt.plot(meta_df['Date'],
         meta_df['High'],
         marker = 'o', 
         markerfacecolor = "pink",
        markeredgecolor = "black",
        linestyle = "solid",
        linewidth = 3.5)

plt.xlabel("Date ", fontdict = font_label)
plt.ylabel("Higest value of day", fontdict = font_label)
plt.title("Highest value per day", fontdict= font_title, loc="right")
plt.grid(color ='red', linestyle = 'dotted')
plt.xticks(rotation = 45 , ha = 'right')
plt.show()
```

# Using MPL Finance

```python
meta_df['Date'] = meta_df['Date'].apply(mpl_dates.date2num)
```

```python
meta_df = meta_df.astype(float)
```

```python
fig, ax = plt.subplots(figsize= (10,10))
candlestick_ohlc(ax, meta_df.values, width = 0.5, colorup = 'green', colordown = 'red', alpha = 0.8)

ax.set_xlabel('Dates')
ax.set_ylabel('Price')
fig.suptitle("Stock values of 50 days")

date_format = mpl_dates.DateFormatter('%d-%m-%Y') 
ax.xaxis.set_major_formatter(date_format) 
fig.autofmt_xdate()

fig.tight_layout() 
  
plt.show() 
```
