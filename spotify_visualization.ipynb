```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
presentation_files_path = "./Spotify/datasets/presentation"
processed_files_path = "./Spotify/datasets/processed"
```

```python
tracks_df = pd.read_csv(f"{processed_files_path}/processed_tracks.csv")
feautres_df = pd.read_csv(f"{processed_files_path}/processed_features.csv")
```

```python
top_artists_by_popularity = pd.read_csv(f"{presentation_files_path}/top_artists_by_popularity.csv")
top_tracks_by_features = pd.read_csv(f"{presentation_files_path}/top_tracks_by_features.csv")
top_tracks_by_popularity = pd.read_csv(f"{presentation_files_path}/top_tracks_by_popularity.csv")
genre_group_by_count = pd.read_csv(f"{presentation_files_path}/genre_group_by_count.csv")
yearly_tracks_released_count = pd.read_csv(f"{presentation_files_path}/yearly_tracks_released_count.csv")
avg_dur_track = pd.read_csv(f"{presentation_files_path}/avg_dur_track.csv")
top_artists_by_followers = pd.read_csv(f"{presentation_files_path}/top_artists_by_followers.csv")
mean_popularity_by_year = pd.read_csv(f"{presentation_files_path}/mean_popularity_by_year.csv")
```

```python
font_title = {"family":"serif","size":20, "color":"blue"}
font_label = {"family":"serif", "size":15, "color":"darkred"}
```

# 1. Line Plot

```python
plt.plot(avg_dur_track['year'],
         avg_dur_track['mean_minutes'],
         marker = 'o', 
         markerfacecolor = "red",
        markeredgecolor = "black",
        linestyle = "solid",
        linewidth = 3.5)
plt.xlabel("Year", fontdict = font_label)
plt.ylabel("Avg time of track (in minutes)", fontdict = font_label)
plt.title("Avg time  vs Year", fontdict= font_title, loc="right")
plt.grid(color ='red', linestyle = 'dotted')
plt.show()
```

# 2. Scatter Plot

```python
plt.scatter(yearly_tracks_released_count['year'],
            yearly_tracks_released_count['count'],
            marker = 'o',
            facecolor = "red",
        edgecolor = "black",      
         )

plt.xlabel('Year', fontdict=font_label)
plt.ylabel("number of songs realeased", fontdict= font_label)
plt.title("Number of songs released ", fontdict= font_title, loc = "left")
plt.grid(color="aqua")
plt.show()
```

# 3.Histogram

```python
plt.hist( tracks_df['danceability'],
        color = "maroon")

plt.title("Dance Ability Histogram", fontdict= font_title)
plt.xlabel("danceability")

plt.grid()
plt.show()
```

```python
plt.hist(tracks_df['loudness'],
        color = "darkred")

plt.title("Loundness histo", fontdict= font_title)
plt.xlabel("loudness", fontdict= font_label)
plt.grid(color="skyblue")
plt.show()
```

# 4. Bar Chart

```python
plt.bar(top_artists_by_followers['name'],top_artists_by_followers['followers'],
       color = 'hotpink')

plt.title(" Top  10 Artists with its number of followers", fontdict=font_title)
plt.xlabel("Artists names", fontdict=font_label)
plt.ylabel("number of followers", fontdict= font_label)

plt.xticks(rotation = 45)
# plt.legend()
# plt.grid(color = 'skyblue')
plt.show()
```

# 5. BOX Plot

```python
plt.boxplot(yearly_tracks_released_count['year'], 
            patch_artist=True, 
            notch =True,
            vert=True,
            widths=0.1)
plt.show()
```

```python
plt.boxplot(avg_dur_track['mean_minutes'], 
            vert=True,
           patch_artist=True,
           notch = True)
plt.title("Mean minutes of the tracks ", fontdict= font_title)

plt.grid()
plt.show()
```

# 6.HeatMap

```python
new_df = feautres_df.iloc[:10, :]

data = new_df [['tempo','valence','acousticness','speechiness','liveness','loudness','energy','duration_ms']]


plt.figure(figsize=(10,6))

sns.heatmap(data, annot=True)
```

# 7.Pie Chart

```python
label_genre_list = genre_group_by_count['genre'].tolist()
value_count = genre_group_by_count['count'].tolist()

plt.pie(value_count, labels=label_genre_list, startangle=0, autopct="%1.1f%%") #shadow=True

# plt.legend(title = "Genre wise count of song", loc = "lower right", ncol = 3)
plt.figure(figsize=(10,7))
plt.show()
```

# 8.Violin plot

```python

fig, ax = plt.subplots()
ax.violinplot(tracks_df['loudness'], showmedians=True)

ax.set_title("Loudness analyasis")
ax.set_xlabel("Loudness")
ax.set_ylabel("Values")
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,6))
ax.violinplot(tracks_df['instrumentalness'], showmedians=True)

ax.set_title("instrumentalness analysis")
ax.set_ylabel("Values")
ax.set_xlabel("Instrumentalness")

plt.show()
```

# 9. Area Plot

```python
x = mean_popularity_by_year['year']
y = mean_popularity_by_year['mean_popularity']
```

```python
plt.fill_between(x, y , color ='black', alpha = 0.9, edgecolor ='red', linewidth =2)

plt.xlabel('years', fontdict= font_label)
plt.ylabel('mean popularity of the year', fontdict= font_label)
plt.title("Mean popularity  By year", fontdict= font_title)

plt.grid(color ='skyblue')
plt.figure(figsize=(10,10))
plt.show()
```

# 10. stem plot (lolipop charts)

```python
fig, ax = plt.subplots(figsize = (10,6))

ax.stem(mean_popularity_by_year['year'],mean_popularity_by_year['mean_popularity'], bottom=10) #orientation="horizontal"
```

```python
fig, ax = plt.subplots(figsize =(10,10))

ax.stem(top_artists_by_followers['name'],top_artists_by_followers['followers'],
       linefmt ="black",
       markerfmt = 'Dr')

ax.set_xlabel("Name of artists")
ax.set_ylabel("Number of followers")
ax.set_title("Top 10 artist with number of followers")

plt.xticks(rotation = 45)
plt.show()
```

# 11. Step Graph

```python
avg_dur_track1 = avg_dur_track.iloc[:10, :]
plt.step(avg_dur_track1['year'], avg_dur_track1['mean_minutes'], ) #'g^'

plt.xlabel('year', fontdict= font_label)
plt.ylabel('AVG min of tracks', fontdict= font_label)
plt.title("Average time per year of the tracks", fontdict=font_title)

# plt.bar(avg_dur_track1['year'], avg_dur_track1['mean_minutes'],)
plt.show()
```

# 12. Pair Plot

```python
import seaborn as sns

tracks_df_slice = tracks_df.iloc[:100, :]

sns.pairplot(tracks_df_slice, vars =['instrumentalness','acousticness','speechiness'], hue ="popularity")
```

# 13. 3D plot

```python
fig = plt.figure(figsize= (10,6))
ax = plt.axes(projection = '3d')

ax.scatter(tracks_df_slice['instrumentalness'],tracks_df_slice['acousticness'],tracks_df_slice['speechiness'])
ax.set_xlabel("instrumentalness")
ax.set_ylabel("acousticness")
ax.set_zlabel("speechiness")
ax.set_title("instrumentalness VS acousticness VS speechiness")
plt.show()
```
